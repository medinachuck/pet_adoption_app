```python
import pandas as pd
import numpy as np
```

```python
# Pets Train CSV import
train_data = ('petfinder_adoption_data/train.csv')
df_train_data = pd.read_csv(train_data)
df_train_data.head()
```

```python
print(df_train_data.count())
print(df_train_data.isnull().values.any())
```

```python
df_train_data.columns
```

```python
df_train_data.dtypes
```

```python
## Rename columns for Merger
df_train_data = df_train_data.rename(columns={"Breed1": "BreedID", "Breed2": "BreedID2", "Color1": "ColorID", "Color2": "ColorID2", "Color3": "ColorID3", "State":"StateID"})
```

```python
df_train_data
```

```python
# Pets Breed Labels CSV import
breed_labels = ('petfinder_adoption_data/breed_labels.csv')
df_breed_labels = pd.read_csv(breed_labels)
df_breed_labels.head()
```

```python
print(df_breed_labels.count())
print(df_breed_labels.isnull().values.any())
```

```python
df_breed_labels.columns
```

```python
df_breed_labels['BreedID'] = df_breed_labels['BreedID'].astype(int)
```

```python
df_breed_labels.dtypes
```

```python
# Pets Color Labels CSV import
color_labels = ('petfinder_adoption_data/color_labels.csv')
df_color_labels = pd.read_csv(color_labels)
df_color_labels.head()
```

```python
print(df_color_labels.count())
print(df_color_labels.isnull().values.any())
```

```python
df_color_labels.columns
```

```python
# Pets State Labels CSV import
state_labels = ('petfinder_adoption_data/state_labels.csv')
df_state_labels = pd.read_csv(state_labels)
df_state_labels.head()
```

```python
print(df_state_labels.count())
print(df_state_labels.isnull().values.any())
```

```python
df_state_labels.columns
```

```python
# Combine Dataframes
# result = pd.concat([df1, df4], axis=1, join="inner")

# Try to join Breed1 from Train.csv with BreedID from breed_labels.csv

df_train_data = pd.merge(df_train_data, df_breed_labels, on="BreedID", how="left")
df_train_data = pd.merge(df_train_data, df_color_labels, on="ColorID", how="left")
df_train_data = pd.merge(df_train_data, df_state_labels, on="StateID", how="left")
```

```python
df_train_data
```

```python
df_train_data = df_train_data.rename(columns={"BreedName": "PrimaryBreed", "ColorName": "PrimaryColor"})
df_train_data
```

```python
df_country = pd.DataFrame(
    {"StateName": ["Johor", "Kedah", "Kelantan", "Kuala Lumpur", "Labuan", "Melaka", "Negeri Sembilan", "Pahang", "Perak", "Perlis", "Pulau Pinang", "Sabah", "Sarawak", "Selangor", "Terengganu"],
     "Country": ["Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia","Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia", "Malaysia"]
     })

df_country
```

```python
df_train_data = pd.merge(df_train_data, df_country, on="StateName", how="left")
```

```python
# Gender - Gender of pet (1 = Male, 2 = Female, 3 = Mixed, if profile represents group of pets)
# df['year']=df['year'].astype(int)

df_gender = pd.DataFrame(
    {"Gender": ["1", "2", "3"],
     "GenderofPet": ["Male", "Female", "Mixed"]
     })

df_gender['Gender']= df_gender['Gender'].astype(str)
df_train_data['Gender']= df_train_data['Gender'].astype(str)
```

```python
df_train_data = pd.merge(df_train_data, df_gender, on="Gender", how="right")
```
